## **Extração de Características em Imagens**

```python
## Importando as libs
from skimage import data
from skimage.color import rgb2gray
from skimage.transform import resize
import matplotlib.pyplot as plt
```

```python
# Carregamento da imagem
img = data.astronaut()

# Redimensionando a imagem
img = resize(img, (256, 256), anti_aliasing=True)
plt.imshow(img)
```

```python
## Transformação da Imagem em padrão Grayscale
import numpy as np

gray = rgb2gray(img)
gray_uint8 = (gray * 255).astype(np.uint8)

plt.imshow(gray, cmap='gray')
plt.title('Imagem em escala de cinza')
plt.axis('off')
plt.show()
```

```python
# Local Binary Patterns
from skimage.feature import local_binary_pattern

## Suprimindo Warnings
import warnings
warnings.filterwarnings('ignore')

# Raio da vizinhança circular. Os vizinhos estarão a uma distância de 3 pixels do centro.
radius = 3

# Número de vizinhos igualmente espaçados ao longo do círculo.
# 8 * radius = 24 significa que você terá 24 vizinhos
n_points = 8 * radius

lbp = local_binary_pattern(gray, n_points, radius, method='uniform')

plt.imshow(lbp)
plt.title('LBP')
plt.axis('off')
plt.show()
```
